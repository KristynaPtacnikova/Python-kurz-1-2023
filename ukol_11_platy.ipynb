```python
import pandas
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/lutydlitatova/python-jaro-2022/main/ukoly/data/platy_2021_02.csv"
platy = pandas.read_csv(url)
print(platy)

platy.plot()
plt.show()
```

```python
platy.plot(kind='bar', color='yellow', grid=True,figsize=(10, 5))
plt.show()
```

Histogram:

```python
platy.plot(kind='hist', bins=5, grid=True, subplots=True)
```
