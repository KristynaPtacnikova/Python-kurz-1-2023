```python
import pandas
```

```python
zamestnanci_praha = pandas.read_csv('zam_praha.csv')
zamestnanci_praha
```

```python
zamestnanci_liberec = pandas.read_csv('zam_liberec.csv')
zamestnanci_liberec
```

```python
zamestnanci_plzen = pandas.read_csv('zam_plzeň.csv')
zamestnanci_plzen
```

```python
zamestnanci_praha['výkon_prace'] = 'zam_praha'
zamestnanci_liberec['výkon_prace'] = 'zam_liberec'
zamestnanci_plzen['výkon_prace'] = 'zam_plzeň'
zamestnanci = pandas.concat([zamestnanci_praha,zamestnanci_liberec,zamestnanci_plzen], ignore_index=True)
zamestnanci
```

```python
platy = pandas.read_csv('platy_2021_02.csv')
platy
```

```python
novy_propojeny_df = pandas.merge(platy, zamestnanci, on=['cislo_zamestnance'], how="outer")
novy_propojeny_df
```

```python
print(novy_propojeny_df.shape)
```

```python
novy_propojeny_df.groupby("výkon_prace")["plat"].mean()
```

```python
vykazy = pandas.read_csv('vykazy.csv')
vykazy
```

```python
vykazy.rename(columns={'emloyee_id':'cislo_zamestnance'})
vykazy
```

```python
novy_propojeny_df_a_vykazy = pandas.merge(novy_propojeny_df,vykazy, on=['cislo_zamestnance'], how="outer")
novy_propojeny_df_a_vykazy
```
